```python
import os
import pandas as pd

# Read the dataset from the CSV file
file_path = "D:/Alan/Python/WS IMDB/Dataset/ForrestGump_reviews.csv"
df = pd.read_csv(file_path)

# Separate data based on the "Rating Stars"
mask_na = df["Stars"].isna()
data_test = df[mask_na]
data_train = df[~mask_na]

# Create the "Dataset" directory if it doesn't exist
directory_path = "D:/Alan/Python/WS IMDB/Dataset/Splited_Dataset"
if not os.path.exists(directory_path):
    os.makedirs(directory_path)

# Save data to CSV files
data_test.to_csv(os.path.join(directory_path, "data_val.csv"), index=False)
data_train.to_csv(os.path.join(directory_path, "data.csv"), index=False)

print("Data berhasil dipisahkan dan disimpan dalam file data_val.csv dan data.csv.")
```
